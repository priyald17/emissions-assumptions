# <center> PJM Generation by Fuel: Data Visualization </center> 

```python
import os
import pandas as pd
```

```python
df = pd.read_csv(
    os.path.join(os.pardir, 'formatted_data','pjm_gen_by_fuel_type.csv'), parse_dates=[0], index_col=0)
```

```python
# Mean hourly in all years. Of non-emitting sources, nuclear is most prominent. Wind is a distant second.  
df.reset_index().groupby('FUEL_TYPE').mean().sort_values('PERCENT_TOTAL', ascending=False)
```

```python
# Mean hourly in 2016 
df[df.index.year == 2016].reset_index().groupby('FUEL_TYPE').mean().sort_values('PERCENT_TOTAL', ascending=False)
```

```python
# Mean hourly in 2017
df[df.index.year == 2017].reset_index().groupby('FUEL_TYPE').mean().sort_values('PERCENT_TOTAL', ascending=False)
```
